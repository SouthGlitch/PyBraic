```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import vectors as V
```

```python
ax = plt.axes()
ax.arrow(0,0,2,3, head_width=0.5, head_length=0.5)
plt.xlim(0,5)
plt.ylim(0,5)
plt.show()
```

```python
a = V.Vector([2, 3])
b = 1.5 * a
c = -1 * a
d = -1.5 * a
props = {
    "head_length":.2,
    "head_width":.2,
}
ax = plt.axes()
ax.arrow(0, 0, *b.values, **props, color=mcolors.to_rgba('green', .6))
ax.arrow(0, 0, *a.values, **props, color=mcolors.to_rgba('blue', .7))
ax.arrow(0, 0, *d.values, **props, color=mcolors.to_rgba('black', 1))
ax.arrow(0, 0, *c.values, **props, color=mcolors.to_rgba('darkred', 1))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, layout="constrained")
ax1: plt.Axes = axs[0]
ax2: plt.Axes = axs[1]
props = {
    "head_length":.2,
    "head_width":.2,
}
ax1.arrow(0,0,*b.values, **props, color='#848484')
ax1.arrow(0,0,*a.values, **props, color='#4c4c4c')
ax1.set_ylim(0,5)
ax1.set_xlim(0,5)

plt.xlim(0,5)
plt.ylim(0,5)
plt.show()
```
